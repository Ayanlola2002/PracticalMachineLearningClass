```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("train.csv", 
                      dtype={"Age": np.float64}, )
test    = pd.read_csv("test.csv", 
                      dtype={"Age": np.float64}, )
```

```python
titanic.head()
```

```python
titanic.shape,test.shape
```

```python
titanic = titanic.drop(['PassengerId','Name','Ticket','Cabin'], axis=1)
test = test.drop(['Name','Ticket','Cabin'], axis=1)
```

```python
titanic.head() 
```

```python
titanic.isnull().sum(),test.isnull().sum()
```

```python
titanic.Embarked.mode(),titanic.Fare.mean()
```

```python
titanic.Embarked.fillna('S', inplace=True),test.Fare.fillna(titanic.Fare.mean(), inplace=True)
titanic.isnull().sum(),test.isnull().sum()
```

```python
titanic.Age.fillna(titanic.Age.mean(), inplace=True),test.Age.fillna(titanic.Age.mean(), inplace=True)
titanic.isnull().sum(),test.isnull().sum()
```

```python
titanic['Sex_Female'] = titanic.Sex.map({'male':0, 'female':1})
test['Sex_Female'] = test.Sex.map({'male':0, 'female':1})
```

```python
embarked_dummies = pd.get_dummies(titanic.Embarked, prefix='Embarked')
embarked_dummies.drop(embarked_dummies.columns[0], axis=1, inplace=True)

titanic = pd.concat([titanic, embarked_dummies], axis=1)
titanic.head()
```

```python
embarked_dummies = pd.get_dummies(test.Embarked, prefix='Embarked')
embarked_dummies.drop(embarked_dummies.columns[0], axis=1, inplace=True)

test = pd.concat([test, embarked_dummies], axis=1)
test.head()
```

```python
titanic['family']=titanic['SibSp']+titanic['Parch']
titanic.head(10)
```

```python
test['family']=test['SibSp']+test['Parch']
test.head(10)
```

```python
titanic['Fare_cuad']=titanic['Fare']**2
titanic['inv_Age']=1/titanic['Age']
titanic.head(10)
```

```python
test['Fare_cuad']=test['Fare']**2
test['inv_Age']=1/test['Age']
test.head(10)
```

```python
per_80=np.percentile(titanic.Fare, 80)
per_80_2=np.percentile(test.Fare, 80)
print(per_80),print(per_80_2)
```

```python
filtro=(titanic['Fare'] > per_80) & (titanic['Pclass'] == 1)
filtro_test=(test['Fare'] > per_80_2) & (test['Pclass'] == 1)
```

```python
titanic['Fare_Pclass_High'] = 0
test['Fare_Pclass_High'] = 0
```

```python
titanic.loc[filtro,'Fare_Pclass_High'] = 1
test.loc[filtro_test,'Fare_Pclass_High'] = 1
```

```python
filtro2=(titanic['Fare'] > per_80) & (titanic['Embarked'] == 'S')
titanic['Embarked_S_high'] = 0
titanic.loc[filtro2,'Embarked_S_high'] = 1

filtro2_test=(test['Fare'] > per_80_2) & (test['Embarked'] == 'S')
test['Embarked_S_high'] = 0
test.loc[filtro2_test,'Embarked_S_high'] = 1
```

```python
filtro3=(titanic['Fare'] > per_80) & (titanic['Embarked'] == 'Q')
titanic['Embarked_Q_high'] = 0
titanic.loc[filtro3,'Embarked_Q_high'] = 1

filtro3_test=(test['Fare'] > per_80_2) & (test['Embarked'] == 'Q')
test['Embarked_Q_high'] = 0
test.loc[filtro3_test,'Embarked_Q_high'] = 1
```

```python
filtro4=(titanic['Fare'] > per_80) & (titanic['Embarked'] == 'C')
titanic['Embarked_C_high'] = 0
titanic.loc[filtro4,'Embarked_C_high'] = 1

filtro4_test=(test['Fare'] > per_80_2) & (test['Embarked'] == 'C')
test['Embarked_C_high'] = 0
test.loc[filtro4_test,'Embarked_C_high'] = 1
```

```python
filtro5=(titanic['Age'] > 50) & (titanic['Sex'] == 'female')
titanic['Age_Female_high'] = 0
titanic.loc[filtro5,'Age_Female_high'] = 1

filtro5_test=(test['Age'] > 50) & (test['Sex'] == 'female')
test['Age_Female_high'] = 0
test.loc[filtro5_test,'Age_Female_high'] = 1
```

```python
filtro6=(titanic['Age'] > 50) & (titanic['Sex'] == 'male')
titanic['Age_male_high'] = 0
titanic.loc[filtro6,'Age_male_high'] = 1

filtro6_test=(test['Age'] > 50) & (test['Sex'] == 'male')
test['Age_male_high'] = 0
test.loc[filtro6_test,'Age_male_high'] = 1
```

```python
filtro7=(titanic['Age'] < 15) & (titanic['Sex'] == 'female')
titanic['Age_female_low'] = 0
titanic.loc[filtro7,'Age_female_low'] = 1

filtro7_test=(test['Age'] < 15) & (test['Sex'] == 'female')
test['Age_female_low'] = 0
test.loc[filtro7_test,'Age_female_low'] = 1
```

```python
filtro8=(titanic['Age'] < 15) & (titanic['Sex'] == 'male')
titanic['Age_male_low'] = 0
titanic.loc[filtro8,'Age_male_low'] = 1

filtro8_test=(test['Age'] < 15) & (test['Sex'] == 'male')
test['Age_male_low'] = 0
test.loc[filtro8_test,'Age_male_low'] = 1
```

```python
feature_cols = ['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_Female', 'Embarked_Q', 'Embarked_S', 'family',
'Fare_cuad', 'inv_Age', 'Fare_Pclass_High', 'Embarked_S_high',
'Embarked_Q_high', 'Embarked_C_high', 'Age_Female_high',
'Age_male_high', 'Age_female_low', 'Age_male_low']
X_train = titanic[feature_cols]
Y_train = titanic.Survived
X_test = test[feature_cols]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.cross_validation import cross_val_score
from sklearn.linear_model import LogisticRegression

selection_method = []
for i in range(1,23):
    logreg = LogisticRegression(C=1e9)
    selection_method = SelectKBest(k=i)
    selection_method.fit(X_train, Y_train)
    X_selection_method = selection_method.transform(X_train)
    print(i,pd.Series(cross_val_score(logreg, X_selection_method, Y_train, cv=10, scoring='accuracy')).mean())
```

```python
selection_method = SelectKBest(k=16)
selection_method.fit(X_train, Y_train)
X_selection_method = selection_method.transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1e9)
reg = logreg.fit(X_selection_method, Y_train)
reg.coef_
```

```python
# make predictions for testing set
X_selection_method = selection_method.transform(X_test)
y_pred_class = logreg.predict(X_selection_method)
```

```python
submission = pd.DataFrame({'PassengerId': test['PassengerId'],
                            'Survived': y_pred_class })
submission.to_csv("titanic_submission_k8.csv", index=False)
```
